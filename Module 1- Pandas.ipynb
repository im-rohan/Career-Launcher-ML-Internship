### Query 1.5
Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
# function 1
def avg_price(N):
    avg=renewed['Average Price'].tail(N).sum()/N
    return avg
print('Average price over the last 1 week ', avg_price(5))
print('Average price over the last 2 weeks', avg_price(10))
print('Average price over the last 1 month', avg_price(21))
print('Average price over the last 3 months', avg_price(63))
print('Average price over the last 6 months', avg_price(126))
print('Average price over the last 1 year', avg_price(252))
```

```python
# function 2
def avg_per(N):
    diff=renewed['Close Price'].tail(N).iloc[N-1]-renewed['Close Price'].tail(N).iloc[0]
    if diff>0:
        profit=(diff*100)/renewed['Close Price'].tail(N).iloc[N-1]
        return profit
    elif diff<0:
        loss=((-(diff))*100)/renewed['Close Price'].tail(N).iloc[N-1]
        return loss
    else:
        return 0
print('Profit/Loss percentage between 1 week', avg_per(5))
print('Profit/Loss percentage between 2 weeks', avg_per(10))
print('Profit/Loss percentage between 1 month', avg_per(21))
print('Profit/Loss percentage between 3 months', avg_per(63))
print('Profit/Loss percentage between 6 months', avg_per(126))
print('Profit/Loss percentage between 1 year', avg_per(252))
```

### Query 1.7
Add another column 'Trend' whose values are:
'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1
'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
'Positive' for 'Day_Perc_Change' in between 1 and 3
'Negative' for 'Day_Perc_Change' in between -1 and -3
'Among top gainers' for 'Day_Perc_Change' in between 3 and 7
'Among top losers' for 'Day_Perc_Change' in between -3 and -7
'Bull run' for 'Day_Perc_Change' >7
'Bear drop' for 'Day_Perc_Change' <-7

```python
def check (renewed):
    if ((renewed['Day_Per_Change']>=(-0.5))& (renewed['Day_Per_Change']<=0.5)) :
        return 'Slight or No change'
    elif ((renewed['Day_Per_Change']>=0.5) & (renewed['Day_Per_Change']<=1)) :
        return 'Slight positive' 
    elif ((renewed['Day_Per_Change']>=(-1)) & (renewed['Day_Per_Change']<=(-0.5))):
        return 'Slight negative'
    elif ((renewed['Day_Per_Change']>=1) & (renewed['Day_Per_Change']<=3)) :
        return 'Positive'
    elif ((renewed['Day_Per_Change']>=(-3)) & (renewed['Day_Per_Change']<=(-1))) :
        return 'Negative'
    elif ((renewed['Day_Per_Change']>=3) & (renewed['Day_Per_Change']<=7)) :
        return 'Among top gainers'
    elif ((renewed['Day_Per_Change']>=(-7)) & (renewed['Day_Per_Change']<=(-3))) :
        return 'Among top losers'
    elif (renewed['Day_Per_Change']>7) :
        return 'Bull run'
    elif (renewed['Day_Per_Change']<(-7)) :
        return 'Bear drop'
    else:
        return 'undetermined'
    
renewed['Trend']=renewed.apply(check,axis=1)
```
