### Query 6.2
Calculate average annual percentage return and volatility of all 30 stocks over a theoretical one year period

```python
stock=stock.dropna()
stock.shape
```

```python
data=pd.DataFrame()
data['Annual Returns']=(stock.pct_change()).mean()*252
data['Volatility']=(stock.pct_change()).std()*(np.sqrt(252))
data
```

### Query 6.3
Cluster the 30 stocks according to their mean annual Volatilities and Returns using K-means clustering. Identify the optimum number of clusters using the Elbow curve method

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 1)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### Query 6.4
Prepare a separate Data frame to show which stocks belong to the same cluster

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 1)
kmeans.fit(data)
```

```python
data['cluster']=kmeans.predict(data)
data
```

```python
import seaborn as sns
plt.figure(figsize=(20,10))
sns.lmplot(data=data,x='Annual Returns',y='Volatility',hue='cluster',fit_reg=False,legend=False)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],c='red',marker='x',label='Centroid')
for i, txt in enumerate(data.index):
    plt.annotate(txt, (data['Annual Returns'][i], data.Volatility[i]))
plt.grid(True)
plt.legend()
plt.show()
```

From the above graph we can see that most stocks are in Cluster 0 and thus, it is not possible to display the data properly.
To overcome the issue we will view Cluster 0 solely as basic data by removing the stocks in other clusters

```python
copy=data.iloc[:,[0,1]]
copy=copy.drop(['HDFC','MUTHOOTFIN'])
copy
```

```python
kmeans.fit(copy)
copy['cluster']=kmeans.predict(copy)
copy
```

```python
plt.figure(figsize=(20,10))
sns.lmplot(data=copy,x='Annual Returns',y='Volatility',hue='cluster',fit_reg=False,legend=False)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],c='red',marker='x',label='Centroid')
for i, txt in enumerate(copy.index):
    plt.annotate(txt, (copy['Annual Returns'][i], copy.Volatility[i]))
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Similar to previous plot, this plot also has the issue of all stock being in minority space. To visualize better, we'll folow the same procedure of selecting the crowded clusters as basic data

```python
sample=copy.iloc[:,[0,1]]
sample=sample.drop(['DHFL'])
sample
```

```python
kmeans.fit(sample)
sample['cluster']=kmeans.predict(sample)
sample
```

```python
plt.figure(figsize=(20,10))
sns.lmplot(data=sample,x='Annual Returns',y='Volatility',hue='cluster',fit_reg=False,legend=False,palette="Set1")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],c='red',marker='x',label='Centroid')
for i, txt in enumerate(sample.index):
    plt.annotate(txt, (sample['Annual Returns'][i], sample.Volatility[i]))
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
